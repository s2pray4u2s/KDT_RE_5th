### 그래프 구성 요소(1)
    - figure: 여러 그래프가 그려지는 최상위 컨테이너
    - axes: 실제 그래프가 그려지는 공간
    - axis: 그래프의 축

```python
# subplots()
    # 여러개의 그래프 그릴 때 사용
    # figure, axes 동시 반환
    # fig, axs = plt.subplots(행, 열)
fig, axe = plt.subplots() # 하나의 figure와 하나의 axes 만듦
```

```python
# 1행 2열의 axes 생성
fig, axs = plt.subplots(1, 2)
axs[0].plot([1, 2, 3], [1, 2, 3])
axs[1].plot([1, 2, 3], [1, 4, 9])
```

```python
# 2행 2열의 axes 생성
fig, axs = plt.subplots(2, 2)
a = [1, 2, 3]
b = [1, 4, 9]
fig.suptitle("여러 함수 그래프")

axs[0, 0].plot(a, a)
axs[0, 0].set_title("1번 그래프")

axs[0, 1].plot(a, b)
axs[0, 1].set_title("2번 그래프")

axs[1, 0].plot(a, [3, 2, 1])
axs[1, 0].set_title("3번 그래프")

axs[1, 1].plot(a, [9, 4, 1])
axs[1, 1].set_title("4번 그래프")

plt.tight_layout() # 그래프 간격 자동 조절
```

```python
# title
plt.plot([1, 2, 3, 4, 5], [1, 4, 9, 16, 25], label = "제곱 그래프", linestyle = "--", color = "c", marker = "o")
plt.title("제곱 함수 그래프", color = "b", fontsize = 20)

# xlabel, ylabel
plt.xlabel("X 값")
plt.ylabel("Y 값 (X의 제곱)")

# 범례
plt.legend()

# 격자
plt.grid(linestyle = "--")
```

```python
# 하나의 axes
x = np.arange(0, 10)
y1 = 2 * x + 1
y2 = 3 * x - 5

plt.plot(x, y1, marker = "o", label = "y = 2x + 1", color = "c")
plt.plot(x, y2, marker = "^", label = "y = 3x - 5", color = "m")

plt.title("두 개의 선 그래프 비교")
plt.xlabel("X값")
plt.ylabel("y값")
plt.legend()
```

```python
# 실습1
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales_2019 = [100, 120, 140, 110, 130, 150, 160, 170, 180, 200, 190, 210]
sales_2020 = [90, 110, 130, 120, 140, 160, 170, 160, 150, 180, 200, 190]

# title
plt.plot(months, sales_2019, label = "2019", linestyle = "-", color = "c", marker = "o")
plt.plot(months, sales_2020, label = "2020", linestyle = "-", color = "m", marker = "^")
plt.title("Monthly Sales Comparison (2019 - 2020)", color = "b", fontsize = 20)

# xlabel, ylabel
plt.xlabel("Month")
plt.ylabel("Sales")

# 범례
plt.legend()

# 격자
plt.grid(linestyle = "--")

plt.show()
```

```python
# 막대 그래프
labels = ["A",  "B",  "C",  "D"]
values = [10, 24, 36, 18]
plt.bar(labels, values, width = 0.8, color = "skyblue")

plt.title("카테고리별 값")
plt.xlabel("카테고리")
plt.ylabel("값")

plt.grid(axis = "y", linestyle = "--", alpha = 0.5)

plt.show()
```

```python
# 여러 막대 그래프
labels = ["1분기", "2분기", "3분기", "4분기"]
men_means = [20, 34, 30, 35]
women_means = [25, 32, 34, 40]

x = np.arange(len(labels))

fig, ax = plt.subplots()
width = 0.4

ax.bar(x - width/2, men_means, width, label="남성")
ax.bar(x + width/2, women_means, width, label="여성")

ax.set_xticks(x, labels)

plt.show()
```

```python
# 실습2. 막대그래프 그리기

categories = ["Category 1", "Category 2", "Category 3", "Category 4", "Category 5"]
data = [20, 35, 15, 27, 45]

plt.bar(categories, data, width = 0.5, color = "navy")

plt.title("Bar Chart")
plt.xlabel("Categories")
plt.ylabel("values")

plt.grid(alpha = .5)

plt.ylim(0, 50)

plt.xticks(rotation = 45, ha ="right")

plt.show()
```

```python
# 히스토그램

# 랜덤 데이터 생성
data = np.random.randn(1000)

# 히스토그램 그리기
plt.hist(data, bins = 30, color = "navy",  edgecolor = "black", alpha = 0.7)

plt.title("정규분포 데이터의 히스토그램")
plt.xlabel("값")
plt.ylabel("빈도(개수)")

plt.show()
```

```python
# 실습3. 히스토그램 그리기

data = np.random.randint(1, 7, size=100)

bins = np.arange(0.5, 7.5, 1)

plt.figure(figsize=(8, 5))
plt.hist(data, bins=bins, color="skyblue", edgecolor="black", alpha=0.8, width=1)

plt.xticks(range(1, 7))
plt.title("주사위 눈 빈도 히스토그램", fontsize=14)
plt.xlabel("주사위 눈", fontsize=12)
plt.ylabel("빈도", fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.4)

plt.show()
```

```python
# 산점도
x = np.linspace(2, 30, 30)
y = np.linspace(10, 40, 30)

y_noisy = y + np.random.uniform(-7, 7, 30)

plt.scatter(x, y_noisy, color = "navy", marker = "o")

plt.title("linspace + 난수로 불규칙한 산점도")
plt.xlabel("X 값")
plt.ylabel("Y 값")
plt.grid(True, linestyle = "--", alpha = 0.5)

plt.show()
```

```python
# 실습4.  산점도 그리기
x = np.linspace(0, 10, 30)
y = 3 * x + 2 + np.random.normal(0, 3, 30)

plt.scatter(x, y, color = "navy", marker = "o")

plt.title("노이즈가 추가된 산점도")
plt.xlabel("X 값")
plt.ylabel("Y 값")
plt.grid(True, linestyle = "--", alpha = 0.5)

plt.show()
```
