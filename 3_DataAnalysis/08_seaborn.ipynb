```python
# 범주형 데이터 그래프
# countplot
palette = sns.color_palette("pastel")
tips = sns.load_dataset("tips")

plt.figure(figsize=(10, 6))
sns.countplot(data = tips, x = "day", hue = "sex", palette = palette)

plt.title("요일별 방문자 수")
plt.grid(linestyle = "--")
plt.show()
```

```python
# barplot
plt.figure(figsize=(10, 6))
palette = sns.color_palette("colorblind")
sns.barplot(data = tips, x = "day", y = "tip", palette = palette)

plt.title("요일별 평균 팁 금액")
plt.show()
```

```python
# displot()
# 히스토그램 + 커널 밀도(kde)
# kde = True : kde 그래프 그려준다.
sns.displot(data = tips, x = "total_bill", bins = 20)
sns.displot(data = tips, x = "total_bill", kde = True, bins = 20)

plt.title("Total bill")
plt.show()
```

```python
# scatterplot()
# 산점도
plt.figure(figsize=(8, 4))
sns.scatterplot(data = tips, x = "total_bill", y = "tip", hue = "sex")
plt.title("Total bill")
plt.show()
```

```python
# 문제1) 범주형 데이터 그래프
palette = sns.color_palette("pastel")
titanic = sns.load_dataset("titanic")

plt.figure(figsize=(8, 4))
sns.countplot(data = titanic, x = "survived", hue = "sex", palette = palette)

plt.title("타이타닉 생존자 수")
plt.xticks([0, 1], ['사망', '생존'])
plt.grid(linestyle = "--")
plt.show()
```

```python
# 문제2) 바플롯
plt.figure(figsize=(8, 4))
palette = sns.color_palette("deep")
titanic = sns.load_dataset("titanic")
graph2 = sns.barplot(data = titanic, x = "pclass", y = "survived", palette = palette, order = [1, 2, 3])
plt.title("객실 등급별 생존율")
graph2.set_xticklabels(["1등급", "2등급", "3등급"])
plt.show()
```

```python
# 문제3.  히스토그램 + 커널 밀도(kde)
titanic = sns.load_dataset("titanic")
graph3 = sns.displot(data = titanic, x = "age",  hue = "survived", kde = True, stat="probability",  bins = 20)

# 제목
graph3.fig.suptitle("생존자 / 사망자 나이 분포", y=1.05)

# 범례 수정
graph3._legend.set_title("생존 여부")
graph3._legend.texts[0].set_text("사망")
graph3._legend.texts[1].set_text("생존")
plt.show()
```

```python
# 문제4. 산점도
plt.figure(figsize=(8, 4))
titanic = sns.load_dataset("titanic").dropna(subset=["age", "fare", "survived"])
titanic["survived_label"] = titanic["survived"].map({ 0: "사망",1: "생존"})
graph4 = sns.scatterplot(data = titanic, x = "age", y = "fare", hue = "survived_label", hue_order=["사망", "생존"], alpha = 0.6)
plt.title("나이 VS 요금,  생존 여부")
graph4.legend(title="생존 여부")
plt.show()
```

```python
# heatmap()
# 상관계수행렬의 그래프
penguins = sns.load_dataset("penguins")

# 상관계수 행렬
    # 상관계수: 두 변수 사이의 직선 관계가 얼마나 강한지를 숫자로 표현한 값
    # heatmap 에서 색으로 표시되는 숫자가 상관계수
    # 1: 양의 상관관계(하나가 증가하면 다른 것도 같은 비율로 증가)
    # ex) 키와 몸무게
    # 2. 음의 상관관계(하나가 증가하면 다른 변수는 감소)
    # ex) 공부시간과 실수 개수

plt.figure(figsize=(10, 8))
corr = penguins.corr(numeric_only = True)
sns.heatmap(corr, annot = True, fmt = ".2f", cmap = "inferno")
plt.show()
```

```python
# jointplot()
sns.jointplot(data = penguins, x = "flipper_length_mm", y = "body_mass_g", kind = "scatter")
plt.show()
```

```python
# 실습2

# 문제 1
penguins = sns.load_dataset("penguins")
features = ["bill_length_mm", "bill_depth_mm", "flipper_length_mm", "body_mass_g"]
for sp in penguins["species"].unique():
    data_sp = penguins[penguins["species"] == sp][features].dropna()
    corr = data_sp.corr()
    plt.figure(figsize=(8, 8))
    sns.heatmap(corr, annot = True, fmt = ".2f", cmap = "coolwarm")
    plt.title(f"{sp} 종 신체 치수 상관 관계")
    plt.show()
```

```python
# 문제 2
penguins = sns.load_dataset("penguins")
for sp in penguins["species"].unique():
    sns.jointplot(data = penguins[penguins["species"] == sp], x = "flipper_length_mm", y = "body_mass_g")
    plt.suptitle(f"{sp} 종: 날개 길이 VS 몸무게", y = 1.02)
    plt.show()
```

```python
# jointplot()
penguins = sns.load_dataset("penguins")
plt.figure(figsize=(12, 8))
sns.jointplot(data = penguins, x = "flipper_length_mm", y = "body_mass_g", kind = "scatter", hue = "species")
plt.show()
```
