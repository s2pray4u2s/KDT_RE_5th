```python
import cv2
```

```python
image = cv2.imread('images/Columbina.webp')
cv2.imshow('Original Image', image)

# 이미지 자르기
image_crop = image[100:200, 100:400]
cv2.imshow('Cropped Image', image_crop)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 이미지 불러오기
img2 = cv2.imread('images/Cyrene.jpg')

# 원본 이미지를 가로 세로 2배 축소
img2_small = cv2.pyrDown(img2)

# 축소한 이미지를 좌우 반전
img2_small_flip = cv2.flip(img2_small, 1)

# 이미지를 원본과 동일한 크기의 복사 이미지의 우하단에 배치
img2_copy = img2.copy()
H, W = img2_copy.shape[:2] # 원본 크기
h, w = img2_small_flip.shape[:2] # 축소 이미지 크기

img2_copy[H - h:, W - w:] = img2_small_flip

cv2.imshow('result image', img2_copy)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('images/Cyrene.jpg')

if img is None:
    print("Image load failed!")
else:
    h, W, _ = img.shape

    # 1/2 크기로 축소
    small_img = cv2.resize(img, (W // 2, h // 2))

    # 좌우 반전
    flip_img = cv2.flip(small_img, 1)

    # 우하단 배치
    result = img.copy()
    sh, sw, _ = flip_img.shape

    # 시작점부터 이미지 높이 만큼만 영역 지정
    result[h // 2 : h // 2 + sh, W // 2: W // 2 + sw] = flip_img

    cv2.imshow('result', result)

    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
