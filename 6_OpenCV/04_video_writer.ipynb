```python
import cv2
```

```python
# 원본 영상 불러오기
cap = cv2.VideoCapture('images/video.mp4')

# VideoWriter 설정을 위한 정보 획득
fourcc = cv2.VideoWriter_fourcc(*'H264')           # 코덱 설정
fps = cap.get(cv2.CAP_PROP_FPS)                    # 원본 FPS 가져오기
width = round(cap.get(cv2.CAP_PROP_FRAME_WIDTH))   # 원본 너비 가져오기
height = round(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)) # 원본 높이 가져오기

# VideoWriter 객체 생성
out = cv2.VideoWriter('outputs/video_output.mp4', fourcc, fps, (width, height))

while cap.isOpened():
    ret, frame = cap.read()

    if not ret:
        break

    # 프레임 저장 >> 주의) 이 방식은 Sound 포함 불가
    out.write(frame)

    cv2.imshow('Video Writer', frame)

    # 'q' 키를 누르면 종료
    if cv2.waitKey(int(1000 / fps)) == ord('q'):
        cv2.imwrite('outputs/capture_video.jpg', frame)
        break

# 자원 해제
cap.release()
out.release()
cv2.destroyAllWindows()   
```
