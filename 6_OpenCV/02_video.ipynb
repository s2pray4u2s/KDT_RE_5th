```python
import cv2
import os
```

```python
# 비디오 파일 또는 웹캠 엽결
cap = cv2.VideoCapture("images/video.mp4")  # 0은 기본 웹캠을 의미, 파일 경로를 넣으면 비디오 파일 열림

while cap.isOpened(): # 영상이 정상적으로 열려있는 동안 반복
    ret, frame = cap.read()  # 프레임 읽기

    if not ret:
        break  # 프레임을 더 이상 읽을 수 없으면 종료

    cv2.imshow("Video Player", frame)  # 프레임 출력

    if cv2.waitKey(33) & 0xFF == ord('q'):  # 'q' 키를 누르면 종료
        break

cap.release()  # 비디오 캡처 객체 해제
cv2.destroyAllWindows()  # 모든 OpenCV 창 닫기 
```

```python
capture = cv2.VideoCapture("images/City.mp4")

target_width = 450      
target_height = 800

fps = capture.get(cv2.CAP_PROP_FPS)
print(f"Frames per second: {fps}")

total_frames = capture.get(cv2.CAP_PROP_FRAME_COUNT)
print(f"Total number of frames: {total_frames}")

# FPS가 0으로 읽히는 예외 상황 방지
wait_time = max(1, int(1000 // fps)) if fps > 0 else 30

# 창의 이름을 미리 정의하고 크기 조절 가능하게 설정
cv2.namedWindow("VideoFrame", cv2.WINDOW_NORMAL) 

# 창을 화면 왼쪽 상단 (100, 100) 좌표로 강제 이동
cv2.moveWindow("VideoFrame", 100, 100)

# 재생 속도 설정을 위한 변수 (기본값)
delay = int(1000 / fps)

while True:

    # 마지막 프레임이면 다시 처음으로
    if capture.get(cv2.CAP_PROP_POS_FRAMES) >= total_frames - 1:
        capture.set(cv2.CAP_PROP_POS_FRAMES, 0)

    ret, frame = capture.read()

    # 영상을 다 읽었거나 읽기에 실패하면 종료
    if not ret:
        break

    # 읽어온 프레임의 크기 조정
    resized_frame = cv2.resize(frame, (target_width, target_height), interpolation = cv2.INTER_AREA)

    # 조절된 프레임을 화면에 표시
    cv2.imshow("VideoFrame", resized_frame)

    # 사용자가 키를 누르면 종료 (q 또는 ESC)
    key = cv2.waitKey(wait_time * 2) 
    if key == ord('q') or key == 27:
        print("User requested exit.")
        break

capture.release()
cv2.destroyAllWindows()
```
