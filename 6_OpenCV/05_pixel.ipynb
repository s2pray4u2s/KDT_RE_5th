```python
import cv2
import numpy as np
```

```python
# 가로 640, 세로 480 크기의 검은색 빈 화면 생성
img = np.zeros((480, 640, 3), dtype = np.uint8)

# 일부 영역 (세로 100~200, 가로 200~300)을 흰색으로 채우기
# BGR 순서로 색상을 지정하므로 (255, 255, 255)은 흰색을 의미
img[100:200, 200:300] = (255, 255, 255)

cv2.imshow('Drawing', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 이미지 채널 분리
image = cv2.imread('images/rgb.png')

# BGR 채널 분리
b, g, r = cv2.split(image)

# 각 채널 화면에 출력
cv2.imshow('Blue Channel', b)
cv2.imshow('Green Channel', g)
cv2.imshow('Red Channel', r)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 분리된 채널을 리스트로 묶어 병합
merged_image = cv2.merge([b, g, r])
cv2.imshow('Merged Image', merged_image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 색상 공간 변환 (BGR -> Grayscale)
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imshow('Grayscale Image', gray_image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 이미지를 흑백으로 변환하여 출력
img2 = cv2.imread('images/Sparxie.webp')
gray_img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
cv2.imshow('Grayscale Image', gray_img2)

# 이미지의 흑백 반전
inverted_img = cv2.bitwise_not(gray_img2)
cv2.imshow('Inverted Image', inverted_img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
