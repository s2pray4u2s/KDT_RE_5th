```python
import cv2
import os
import sys
```

```python
# 이미지 읽기
img = cv2.imread("images/Columbina.webp")

# 이미지 창에 표시
cv2.imshow("Image Window", img)

# 키보드 입력 대기 (0은 무한 대기)
key = cv2.waitKey(0)
print(f"Pressed key code: {key}")

# 모든 창 닫기  
cv2.destroyAllWindows()
```

```python
# 채널 3 (R, G, B) 추출
img_default = cv2.imread("images/Columbina.webp", cv2.IMREAD_COLOR)

# 채널 없음 (그레이스케일)
img_grayscale = cv2.imread("images/Columbina.webp", cv2.IMREAD_GRAYSCALE)

cv2.imshow("Default Image", img_default)
cv2.imshow("Grayscale Image", img_grayscale)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
base_path = os.getcwd() 
logo_path = os.path.join(base_path, "images", "logo.png")

# 채널 3 (R, G, B) 추출
img_default = cv2.imread(logo_path, cv2.IMREAD_COLOR)

# 채널 4 (Alpha 값 포함 - 배경이 투명한 이미지)
img_unchanged = cv2.imread(logo_path, cv2.IMREAD_UNCHANGED)

cv2.imshow("Default Image", img_default)
cv2.imshow("Unchanged Image", img_unchanged)
print("Logo shape (unchanged):", img_unchanged.shape)  # (height, width, channels) 
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 체널 3 (R, G, B) 추출
img_rgb = cv2.imread("images/Castorice.jpg", cv2.IMREAD_COLOR)

# 채널 없음 (그레이스케일)
img_black = cv2.imread("images/Castorice.jpg", cv2.IMREAD_GRAYSCALE)

# 이미지가 제대로 읽혔는지 확인
if img_rgb is None or img_black is None:
    print("이미지 파일을 찾을 수 없습니다. 경로를 확인해주세요.")

else:
    while True:
        
        # [추가] 화면에 이미지 표시 (창 제목, 표시할 이미지)
        cv2.imshow("RGB Color", img_rgb)
        cv2.imshow("Grayscale", img_black)

        # 1ms 마다 입력 대기
        key = cv2.waitKey(1) & 0xFF

        # q 키를 누르면 종료
        if key == ord('q'):
            print("사용자가 q 키를 눌러 종료합니다.")
            break

        # ESC 키(아스키코드 27)를 누르면 종료
        elif key == 27:
            print("사용자가 ESC 키를 눌러 종료합니다.")
            break

cv2.destroyAllWindows()
```
