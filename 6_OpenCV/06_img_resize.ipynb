```python
import cv2
```

```python
# 이미지 불러오기
image = cv2.imread('images/Sparxie.webp')

# 고정 크기로 이미지 리사이즈 (가로 320, 세로 240)
dst_fixed = cv2.resize(image, (320, 240))
cv2.imshow('Fixed Size', dst_fixed)

# 비율로 조정 (가로 50%, 세로 50%)
dst_ratio = cv2.resize(image, None, fx = 0.5, fy = 0.5)
cv2.imshow('Ratio Size', dst_ratio)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 비디오 불러오기
cap2 = cv2.VideoCapture('images/video.mp4')

# 비디오의 원본 크기 및 FPS 정보 가져오기
width = cap2.get(cv2.CAP_PROP_FRAME_WIDTH)
height = cap2.get(cv2.CAP_PROP_FRAME_HEIGHT)
fps = cap2.get(cv2.CAP_PROP_FPS)

# 비디오 리사이즈 수치 결정 (1.5배)
target_width = int(width * 1.5)
target_height = int(height * 1.5)

# 재생 속도 계산
wait_time = max(1, int(1000 // fps)) if fps > 0 else 30

print(f'Original Size: {int(width)} x {int(height)}, FPS: {fps}, Wait Time: {wait_time} ms')
print(f'Target Size: {target_width} x {target_height}')

while True:
    ret, frame = cap2.read()
    if not ret:
        break

    # 비디오 리사이즈
    resized_frame = cv2.resize(frame, (target_width, target_height), interpolation = cv2.INTER_CUBIC)

    # 리사이즈된 프레임 표시
    cv2.imshow('Resized Video', resized_frame)

    # 종료 조건
    if cv2.waitKey(wait_time) & 0xFF == ord('q'):
        break

cap2.release()
cv2.destroyAllWindows()q
```
