### 문제1

### 문제 5
- 두 텐서 a = [1, 2]와 b = [3, 4]를 쌓아 2x2 텐서를 생성해서 출력하세요.

### 문제9
- 텐서 [[1, 2], [3, 4], [5, 6]]에서:
- dim=1 기준으로 크기 1씩 나누세요.
- 각각의 조각을 새로운 축에 쌓은 후 출력해보세요.

````python
# 여기에 코드를 작성합니다
X = torch.tensor([[1, 2], 
                  [3, 4], 
                  [5, 6]])

# 1. 열 기준으로 크기 1씩 나누기
X9_1 = torch.split(X, 1, dim = 1)
print(X9_1)

# 2. 새로운 축에 쌓기
X9_2 = torch.stack(X9_1)
print(X9_2)
````

### 문제10
- 다음 텐서를 활용해서 아래 문제를 해결해보세요
  ```
  a = torch.tensor([[1, 2], [3, 4]])
  b = torch.tensor([[5, 6], [7, 8]])
  ```

1. 두 텐서를 열 기준으로 연결하고 결과를 출력하세요.
2. 연결된 텐서를 크기 `(4, 2)`로 변환하세요.
3. 변환된 텐서를 2개의 조각으로 나누고, 각 조각을 0번 축으로 스택하여 출력하세요.


````python
# 여기에 코드를 작성합니다
a = torch.tensor([[1, 2], 
                  [3, 4]])
b = torch.tensor([[5, 6], 
                  [7, 8]])

# 1. 열 기준으로 병합
X10_1 = torch.cat((a, b), dim = 1)
print(X10_1)

# 2. 크기 (4, 2)로 변환
X10_2 = X10_1.view(4, 2)
print(X10_2)

# 3. 2개의 조각으로 나누고
X10_3 = torch.split(X10_2, 2)
print(X10_3)

# 4. 0번 축으로 스택
X10_4 = torch.stack(X10_3)
print(X10_4)
````
