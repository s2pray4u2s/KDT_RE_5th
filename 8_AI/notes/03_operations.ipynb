## 02. 텐서의 연산

### 2-1. 산술 연산

```python
x = torch.ones(2, 2)
print(x)

x = x * 2
print(x)

x = x + 3
print(x)
```

```python
x = torch.tensor([[1, 2], [3, 4]])
y = torch.tensor([[4, 3], [2, 1]])
```

```python
# 더하기 
add = torch.add(x, y)
print(add)
print(x + y)
```

```python
# 빼기
sub = torch.sub(x, y)
print(sub)
print(x - y)
```

```python
# 곱하기
mul = torch.mul(x, y)
print(mul)
print(x * y)
```

```python
# 나누기
div = torch.div(x, y)
print(div)
print(x / y)
```

### 2-2. 바꿔치기 연산
- 연산자의 뒤에 "_" 를 추가한 형태로 사용

```python
# Tensor 바꿔치기 연산
x = torch.tensor([[5, 6], [7, 8]])
y = torch.tensor([[3, 1], [4, 2]])

# x = torch.add(x, y)
x.add_(y)
print(x)

x.fill_(0)
print(x)
```

### 2-3. 각종 연산

```python
x = torch.tensor([[1.1, 2.2, 3.3], [4, 5, 6], [-1, -3.14, -6]])
print(x)

# 수학 연산
print("절대값", torch.abs(x))
print("올림", torch.ceil(x))
print("반올림", torch.round(x))
print("내림", torch.floor(x))
print("모든 원소의 합", torch.sum(x))
print("모든 원소의 곱", torch.prod(x))
print("클램프", torch.clamp(x, -2, 2))
```

```python
# 통계 연산
print("최솟값", torch.min(x))
print("최댓값", torch.max(x))
print("최솟값의 인덱스", torch.argmin(x))
print("최댓값의 인덱스", torch.argmax(x))
print("평균값", torch.mean(x))
print("표준편차", torch.std(x))
print("중복제거", torch.unique(torch.tensor([1 ,2, 3, 1, 2, 2, 2, 3, 2, 2, 2, 3, 4])))
```

```python
# 차원 지정
x = torch.tensor([[1.1, 2.2, 3.3], [4, 5, 6], [-1, -3.14, -6]])
print("열의 합", torch.sum(x, dim = 0))
print("행의 합", torch.sum(x, dim = 1))
print("열의 최댓값", torch.max(x, dim = 0))
print("행의 최솟값", torch.min(x, dim = 1))
```

### 2-3. 내적(dot product) / 행렬의 곱셈

```python
x = torch.tensor([[1, 2], [3, 4]])
y = torch.tensor([[5, 6], [7, 8]])

print(f"내적의 곱:\n {torch.matmul(x, y)}")
```
