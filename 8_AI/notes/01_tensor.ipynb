# PyTorch
- GPU를 이용한 동적 신경망 구축이 가능한 딥러닝 프레임워크

## PyTorch의 구성요소
- **torch**: 메인 네임스페이스, 텐서 등의 다양한 함수가 포함
- **torch.autograd**: 자동 미분 기능을 제공하는 라이브러리
- **torch.nn**: 신경망 구축을 위한 데이터 구조, 레이어 등으로 제공하는 라이브러리
- **torch.multiprocessing**: 병렬 처리 기능을 제공하는 라이브러리
- **torch.optim**: SGD (확률적 경사 하강법)등 파라미터 최적화 알고리즘 제공
- **torch.utils**: 데이터 조작 등 유틸리티 기능 제공

## 01. 텐서(Tensor)
- 방향성과 크기를 동시에 표현하는 다차원 배열

### 1-1. Tensor의 초기화
- 초기화: 텐서를 생성하고 특정 값을 (최초로) 채우는 것

#### 초기화 되지 않는 텐서

```python
x = torch.empty(4, 2)
print(x)
```

#### 무작위로 초기화된 텐서
- **rand**: 0과 1사이의 값으로 랜덤하개 초기화된 텐서 생성
- **randn**: 표준정규분포를 따르는 랜덤한 값으로 초기화

```python
x = torch.rand(4, 2)
y = torch.randn(2, 3)

print(x)

print(y)
```

#### 0으로 채워진 텐서

```python
x = torch.zeros(4, 2, dtype = torch.long)
print(x)
```

#### 1로 채워진 텐서

```python
x = torch.ones(2, 4, dtype = torch.double)
print(x)

# 기존 텐서의 속성 (데이터 타입, 장치)을 물려받음
y = x.new_ones(3, 2)
print(y)
```

#### 사용자가 입력한 값으로 텐서 초기화

```python
x = torch.tensor([1, 2.5])
print(x)
```

#### numpy에서 불러오기

```python
data = [1, 2, 3, 4]
np_array = np.array(data)
x = torch.from_numpy(np_array)
print(x)
```

#### _like 메서드
- 기존 텐서의 속성을 복사하면서 값만 다른 텐서를 만드는 메서드

```python
x = torch.tensor([[1, 2], [3, 4]], dtype = torch.float64)
y = torch.zeros_like(x)
print(y)
```

### 1-2. 텐서의 속성

```python
t = torch.rand(3, 4)
print(t.size()) # 모양
print(t.shape)  # 모양
print(t.dtype)  # 자료형
print(t.device) # 장치
```

### 1-3. 데이터 타입

#### 특정 데이터 타입의 텐서 생성

```python
x = torch.FloatTensor([1, 2, 3])
print(x.dtype)

y = torch.tensor([1, 2, 3], dtype = torch.float32)
print(y.dtype)
```

#### 타입 캐스팅 (형 변환)

```python
print("======= Float 형변환 =======")
x = torch.FloatTensor([1.1, 2.2, 3.3])
print(x)
print(x.short())
print(x.int())
print(x.long())

print("")

print("======= Int 형변환 =======")
y = torch.IntTensor([1, 2, 3])
print(y)
print(y.half())
print(y.float())
print(y.double())
```

### 1-4. 다차원 텐서 표현

#### 0D Tensor (Scalar)
- 하나의 숫자를 담고 있는 텐서
- 축과 형상이 없음

```python
t0 = torch.tensor(0)
print(t0.ndim)
print(t0.shape)
print(t0)
```

#### 1D Tensor (Vector)
- 값들을 저장한 리스트와 유사한 텐서
- 하나의 축이 존재

```python
t1 = torch.tensor([1, 2, 3])
print(t1.ndim)
print(t1.shape)
print(t1)
```

#### 2D Tensor (Matrix)
- 행렬과 같은 모양으로 두 개의 축이 존재
- 일반적인 수치, 통계 데이터 셋에 주로 활용

```python
t2 = torch.tensor([[1, 2, 3], [1, 2, 3], [1, 2, 3]])
print(t2.ndim)
print(t2.shape)
print(t2)
```

#### 3D Tensor
- 큐브와 같은 모양으로 세 개의 축이 존재
- 데이터가 연속된 시퀀스 데이터나 시간축이 포함된 시계열 데이터에 해당
- EX) 주식 가격, 질병 발병 데이터

```python
t3 = torch.tensor([[[1, 2, 3], [1, 2, 3], [1, 2, 3]], [[1, 2, 3], [1, 2, 3], [1, 2, 3]]])
print(t3.ndim)
print(t3.shape)
print(t3)
```

#### 4D Tensor
- 4개의 축
- 컬러 이미지가 대표적인 사례
- 주로 샘플, 높이, 너비, 컬러 채널을 가진 구조로 사용

```python
t4 = torch.tensor([
    [
        [[1, 2, 3], [1, 2, 3], [1, 2, 3]],
        [[1, 2, 3], [1, 2, 3], [1, 2, 3]]
    ],
    [
        [[1, 2, 3], [1, 2, 3], [1, 2, 3]],
        [[1, 2, 3], [1, 2, 3], [1, 2, 3]]
    ]
])
print(t4.ndim)
print(t4.shape)
print(t4)
```

#### 5D Tensor
- 5개의 축
- 비디오 데이터가 대표적 사례
- 주로 샘플, 프레임, 높이, 너비, 컬러 채널을 가진 구조로 사용
