```python
'''
조건문
- 조건에 따라 프로그램의 실행 흐름을 분기시키는 제어문
- 조건: 참 / 거짓을 구분할 수 있는 문장
'''
```

```python
# 조건문의 기초 문법
# if + 조건 -> 조건이 참이면 실행

a = int(input("숫자를 입력해주세요"))
if a > 10:
    print("a는 10보다 큽니다.")
print("조건문 종료")
```

```python
# 들여쓰기 에러 예시
if a > 10:
    print("조건문 종료")
# print("a는 10보다 큽니다.") >> IndentationError: expected an indented block after 'if' statement on line 2
```

```python
# 조건문에 실행할 코드를 작성하지 않았을 때
# pass 로 해당 조건문을 넘어갈 수 있음.
if a > 100:
    pass
```

```python
# 문제1. 날씨에 따른 준비물 안내.
weather = input("오늘의 날씨를 입력해주세요 (비, 맑음)")
if weather == "맑음":
    print("선크림을 바르세요!")
elif weather == "비":
    print("우산을 챙기세요!")
else:
    print("잘못된 입력입니다.")
```

```python
# if - else 문
# 조건이 참 일때는 if 문을, 조건이 거짓일 때는 else 문을 실행
# else -> ~ 아니라면 의 의미 -> 조건이 필요X, if 문과 반드시 같이 나와야 함.
a = int(input("숫자를 입력해주세요."))
if a > 10:
    print("a는 10보다 큽니다.")
else:
    print("a는 10보다 작습니다.")
```

```python
# 문제2. 짝수, 홀수 판별하기
number = int(input("정수를 입력해주세요."))
if number % 2 == 0:
    print("짝수 입니다.")
else:
    print("홀수 입니다.")
```

```python
# if - elif - else 구문
# elif : else if의 약자, if가 있어야만 사용가능.

score = int(input("점수를 입력해주세요."))
if score >= 90:
    print("등급은 A 입니다.")
elif score >= 80:
    print("등급은 B 입니다.")
elif score >= 70:
    print("등급은 C 입니다.")
elif score >= 60:
    print("등급은 D 입니다.")
else:
    print("등급은 F 입니다.")
```

```python
# 실습3. 나이에 따른 영화 관람 가능 여부
age = int(input("나이를 입력해주세요."))
if age >= 19:
    print("귀하는 청소년 관람불가 가능합니다.")
elif age >= 15:
    print("귀하는 15세 이상 관람 가능합니다.")
elif age >= 12:
    print("귀하는 12세 이상 관람 가능합니다.")
else:
    print("귀하는 전체 관람가 가능합니다.")

```

```python
# 실습4. 시, 분, 초 구하기
watch = int(input("시간을 입력해주세요."))
if watch < 60:
    hour = 0
    minute = 0
    second = watch
elif watch < 3600:
    hour = 0
    minute = watch // 60
    second = watch % 60
else:
    hour = watch // 3600
    remain = watch % 3600
    minute = remain // 60
    second = remain % 60
print(f"{hour}시간 {minute}분 {second}초")

```

```python
# 실습4. 시, 분, 초 구하기 (다른풀이)
hour, minute, second = 0, 0, 0
input_second = int(input("초를 입력해주세요."))

minute = input_second // 60
second = input_second % 60
hour = minute // 60
minute %= 60

if hour > 0:
    print(f"{hour}시간 {minute}분 {second}초")
elif minute >0:
    print(f"{minute}분 {second}초")
else:
    print(f"{second}초")
```

```python
# 중첩 조건문
# 하나의 if문 안에 또 다른 if문을 사용하는 것
username = input("관리자 아이디를 입력해주세요.")
password = input("비밀번호를 입력해주세요.")
if username == "admin":
    if password == "abcd":
        print("로그인이 성공하셨습니다.")
    else:
        print("비밀번호가 틀렸습니다.")
else:
    print("잘못된 사용자 입니다.")
```

```python
# 실습5. 편의점 도시락 구매하기

money = int(input("금액을 입력해주세요."))
if money >= 4000:
    print("편의점 도시락 구매")
elif money >= 2500:
    print("김밥 구매")
elif money >= 1500:
    print("삼각김밥 구매")
else:
    print("금액이 부족합니다.")
```

```python
# 실습5. 편의점 도시락 구매하기
money = int(input("금액을 입력해주세요."))
food = input("식품명을 입력해주세요. (도시락 / 김밥 / 삼각김밥)")
if food == "도시락":
    if money >= 4000:
        print("도시락 구매완료")
    else:
        print("금액이 부족합니다.")
elif food == "김밥":
    if money >= 2500:
        print("김밥 구매완료")
    else:
        print("금액이 부족합니다.")
elif food == "삼각김밥":
    if money >= 1500:
        print("삼각김밥 구매완료")
    else:
        print("금액이 부족합니다.")
else:
    print("잘못된 메뉴 입니다.")
```
