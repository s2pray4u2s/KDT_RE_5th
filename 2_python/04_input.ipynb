```python
'''
사용자입력 (input)
- input함수: 콘솔을 통해 사용자로부터 문자열 형태로 입력 받음
- input함수 사용법
  - my_input = input("콘솔에 띄울 설명")
'''
```

```python
name = input("이름을 입력하세요: ")
print(name)
```

```python
# 숫자로 활용 시 '형 변환' 필요

a = input()
b = input()
print(a + b)
```

```python
a = int(input())
b = int(input())
print(a + b)
```

```python
c = float(input())
d = float(input())
print(c + d)
```

```python
# 여러 자료 입력하기
# 문자열을 쪼개주는 함수: split()
# 기본 구분자: 공백
fruit1, fruit2, fruit3 = input().split()
print(fruit1, fruit2, fruit3)
```

```python
# 다른 구분자 사용
apple, banana, grape = input().split("-")
print(apple, banana, grape)
```

```python
#실습3. input 연습하기
name = input("이름을 입력하세요: ")
age = input("나이를 입력하세요: ")
print(f"안녕하세요. 저는 {name} 이고, {age}살 입니다.")
```

```python
#실습4-1. 입력과 연산 연습하기
width = int(input("가로길이를 입력하세요: "))
height = int(input("세로길이를 입력하세요: "))
area = width * height
perimeter = 2 * (width + height)
print(f"넓이:{area}")
print(f"둘레:{perimeter}")
```

```python
#실습4-2. 입력과 연산 연습하기
thousand = int(input("천의자리를 입력하세요: "))
hundred = int(input("백의자리를 입력하세요: "))
ten = int(input("십의자리를 입력하세요: "))
one = int(input("일의자리를 입력하세요: "))
print(f"천의 자리:{thousand}")
print(f"백의 자리:{hundred}")
print(f"십의 자리:{ten}")
print(f"일의 자리:{one}")
```

```python
#실습5. 발표 순서와 발표 주제 정하기
name1, name2, name3 = input("이름을 입력하세요: ").split()
print(f"발표자: {name1, name2, name3}")
theme1, theme2, theme3 = input("발표주제를 입력하세요: ").split()
print(f"발표주제: {theme1, theme2, theme3}")

print(f"""발표순서 안내 입니다.
1조 발표자: {name1} - 주제: {theme1}
2조 발표자: {name2} - 주제: {theme2}
3조 발표자: {name3} - 주제: {theme3}""")
```

```python
#실습6. 날짜와 시간
year, month, day = input("년, 월, 일을 입력해주세요. ").split(".")
print(f"{year}년 {month}월 {day}일")
hour, minute, second = input("시, 분, 초를 입력해주세요. ").split(":")
print(f"{hour}시 {minute}분 {second}초")
print(f"""RE5의 개강일은 {year}년 {month}월 {day}일
지금 시간은 {hour}시 {minute}분 {second}초 입니다.""")
```
