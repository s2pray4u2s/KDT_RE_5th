```python
'''
set(집합)
- 원소의 중복을 허용하지 않는 여러 데이터의 모음
- 순서가 없는 컬렉션 자료형
'''
```

```python
# set 만들기
s1 = {1, 2, 3}
print(s1, type(s1)) # {1, 2, 3} <class 'set'>

s2 = {1, 1, 2, 3, 4, 5, 6, 6, 7} # {1, 2, 3, 4, 5, 6, 7}
print(s2)
```

```python
# 빈 set 만들기
# - 중괄호에 원소를 넣지 않고 만들면 빈 dict로 인식
s3= {} # <class 'dict'>
print(type(s3))

# set 함수로 생성
s4 = set()
print (s4, type(s4)) # <class 'dict'>

# set 함수의 활용: 원소의 중복 제거
my_list = [1, 1, 2, 2, 2, 2, 2, 3, 3, 3, 4, 4, 4, 4]  # {1, 2, 3, 4}
s5 = set(my_list)
print(s5)

# 인덱싱 X
# s1[1] >> TypeError: 'set' object is not subscriptable

# 자료형 제한
# - 가번 자료형은 원소로 사용할 수 없다
# s1 = {1 ,2 ,3 ,3, [5, 6, 7]} >> TypeError: unhashable type: 'list'
```

```python
# 집합 연산: 합집합, 교집합, 차집합, 대칭차집합
A = {"a", "b", "c"}
B = {"c", "d", "e"}

# 합집합
s_union1 = A | B
s_union2 = A.union(B)
print("합집합1:", s_union1)
print("합집합2:", s_union2)

# 교집합
s_inter1 = A & B
s_inter2 = A.intersection(B)
print("교집합1:", s_inter1)
print("교집합2:", s_inter2)

# 차집합
s_diff1 = A - B
s_diff2 = A.difference(B)
print("차집합1:", s_diff1)
print("차집합2:", s_diff2)

# 대칭 차집합
s_symm1 = A ^ B
s_symm2 = A.symmetric_difference(B)
print("대칭차집합1:", s_symm1)
print("대칭차집합2:", s_symm2)
```

```python
# 실습1. set 종합 연습

# 문제1. 중복 제거 및 개수 세기
submissions = ["kim", "lee", "kim", "park", "choi", "lee", "lee"]
sub1 = set(submissions)
print(f"제출한 학생 수: {len(sub1)} 명")
print(f"제출자 명단: {sub1}")
```

```python
# 문제2. 공통 관심사 찾기
user1 = {"SF", "Action", "Drama"}
user2 = {"Drama", "Romance", "Action"}
print(f"공통 관심 장르: {user1 & user2}")
print(f"서로 다른 장르: {user1 ^ user2}")
print(f"전체 장르: {user1 | user2}")
```

```python
# 문제3. 부분집합 관계 판단
my_certificates = {"SQL", "PYTHON", "LINUX"}
job_required = {"SQL", "PYTHON"}
print(f"공통 자격: {my_certificates & job_required}")
print(f"지원 자격 충족 여부: {my_certificates.issuperset(job_required)}")
```

```python
# set 메서드
s1 = {1, 2, 3}

# 원소 추가
s1.add(4)
print("원소 추가:", s1)

# 여러 원소 추가
s1.update((5, 6, 7))
print("여러 원소 추가:", s1)

# 원소 제거
s1.remove(4)
print("원소제거1:", s1)
# s1.remove(100) >> KeyError: 100

s1.discard(100)
s1.discard(6)
print("원소제거2:", s1)

deleted = s1.pop()
print("원소제거3:", s1, deleted)
```

```python
# 부분 집합 관련 메서드
A = {1, 2}
B = {1, 2, 3}

print(A.issubset(B))
print(B.issuperset(A))
print(A.isdisjoint({4}))
```

```python
A = {10, 20, 30, 40, 50} # 상위 집합
B = {20, 30, 40} # 부분 집합
C = {10, 200, 300, 400, 500}

# 부분집합 여부 판단
print(B.issubset(A)) # True
print(A.issubset(B)) # False

# 상위집합 여부 판단
print(A.issuperset(B)) # True
print(B.issuperset(A)) # False

# 공통 원소가 없는지 확인
print(A.isdisjoint(B)) # False
print(A.isdisjoint(C)) # False
print(B.isdisjoint(C)) # True
```
