```python
# 딕셔너리 반복
person = {"name" : "Won", "age" : 38, "address" : "Seoul"}
for key in person:
    print(f"key: {key}, value: {person[key]}")

# value만 가져오기
for value in person.values():
    print(f"value: {value}")

# items 가져오기
for key, value in person.items():
     print(f"key: {key}, value: {value}")
```

```python
# for문과 range()
# range 함수: 지정된 범위의 정수 시퀀스 생성
# 기본문법
# range(start, end, step)
# - start: 생략 가능. 기본값:0
# - end: end - 1 까지 생성
# - step: 생략 가능. 기본값:1

range(1,5)

for i in range(1,5):
    print(i)

for i in [1, 2, 3, 4]:
    print(i)

# start 생략
for i in range(5):
    print(i)
```

```python
# 간격 (step) 지정
for i in range(0, 11, 2):
    print(i)

for i in range(11, 0, -2):
    print(i)
```

```python
# 루프 제어문
# - 특정 조건 하에서만 작동하도록 구현
# break: 반복을 즉시 중단
for i in range(10):
    if i == 5:
        break
    print(i)
    
print("반복종료")
```

```python
number = int(input("정수를 입력해주세요."))
for i in range(1, number + 1):
    for j in range(number - i):
        print(" ", end = "")
    for j in range(i):
        print("*", end="")
    print()
```

```python
number = int(input("정수를 입력해주세요."))
for i in range(1, number + 1):
    for j in range(number - i):
        print(" ", end = "")
    for j in range(2 * i - 1):
        print("*", end = "")
    print()
```

```python
# 리스트 컴프리헨션
# - for 문을 리스트에 한줄로 축약하여 새 리스트를 생성하는 문법
# - [표현식(리스트의 원소) for 변수 in 반복대상 if 조건]
# - 표현식: 값을 유도하는 식 (표현)

# for문 이용
squares = []
for x in range(1,6):
    squares.append(x ** 2)
print(squares)

# 리스트 컴프리헨션
squares2 = [x ** 2 for x in range(1,6)]
print(squares2)

# 조건문 추가하기
squares3 = [x ** 2 for x in range(1,11) if x % 2 == 0]
print(squares3)
```
