```python
'''
튜플 (tuple)
 - 순서가 존재하는 여러 데이터의 모음
 - 불변(immutable) 자료형
'''
```

```python
# 튜플 생성
my_tuple = (1, 2, 3, 4)
print(my_tuple) # (1, 2, 3, 4)
print(type(my_tuple)) # <class 'tuple'>

my_tuple2 = 5, 6, 7, 8
print(my_tuple2)
print(type(my_tuple2))

# 원소가 하나인 튜플 생성
single_el_tuple = (100,)
print(single_el_tuple)
print(type(single_el_tuple))

# 튜플 생성 함수로 생성
my_tuple2 = tuple()
print(my_tuple2)

my_tuple3 = tuple("CodingOn")
print(my_tuple3)
```

```python
'''
언패킹
- 시퀀스에 지정된 여러 값을 여러 번수에 나눠서 저장하는 것
- 튜플, 리스트, 문자열...
'''

apple, banana, kiwi = ("apple", "banana", "kiwi")
print(apple, banana, kiwi)

# 불변성 (immutable)
# - 객체가 생성된 이후 내부 데이터를 변경할 수 없는 것.

# my_tuple[0] = 100 >> TypeError
# del my_tuple[1] >> TypeError
# del my_tuple
# print(my_tuple) >> NameError
# 튜플 자체는 삭제 가능 (원소 삭제 불가)
```

```python
# 튜플 수정
my_tuple4 = (10, 20, 30)
new_tuple = (100,) + my_tuple4[1:]
print("origin_tuple:", my_tuple4)
print("new_tuple:", new_tuple)
```

```python
# 실습1. 튜플 종합 실습

# step1. 손상된 고객 정보 복원하기
user = ("Minji", 25, "Seoul")
restored_user = ("Eunji",) + user[1:]
print(restored_user)
```

```python
# step2. 고객 정보 언패킹하여 변수에 저장
name, age, city = restored_user
print(name, age, city)
```

```python
# step3. 지역별 보안 정책 분기 거리
```

```python
# step4. 고객 데이터 통계 분석
users = ("Minji", "Eunji", "Soojin", "Minji", "Minji")
print(users.count("Minji"))
print(users.index("Soojin"))
```

```python
# step5. 고객이름 정렬
sorted_users = tuple(sorted(users))
print(sorted_users)
```
