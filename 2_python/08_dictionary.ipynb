```python
'''
딕셔너리
- 키 값 쌍으로 묶어 데이터를 저장하는 자료형
- 키는 유일해야 함. 값은 중복 가능
- 변경 가능한 자료형
- 순서가 보장되지 않았다가 py 3.7 버전 이후로 순서가 보장됨.
'''
```

```python
# dict 만들기
d1 = {} # 빈 딕셔너리 생성
print(d1, type(d1)) # {} <class 'dict'>

person = {"name": "Doohee", "age": 38}
print(person)
```

```python
# dict 함수로 생성
d2 = dict()
print(d2, type(d2)) # {} <class 'dict'>
```

```python
# 키가 문자열일 떄 가능
movie = dict(title = "The Avengers", director = "Joss Whedon")
print(movie)
print(movie, movie["director"])
```

```python
# 리스트나 튜플로 만들기
pairs = [("name", "Doohee"), ("age", 38), ("job", "engineer")]
person2 = dict(pairs)
print(person2)
```

```python
# zip() 함수로 만들기
keys = ["name", "age", "city"]
values = ["Doohee", 38, "Seoul"]
print(dict(zip(keys, values)))
```

```python
# 키로 사용할 수 없는 자료형
# 키 - 불변 자료형만 사용해야한다.
d1 = {(1, 2, 3):(1, 2, 3)} # 튜플 사용 가능
d2 = {1: 10} # 숫자 사용 가능
# d3 = {[1, 2, 3]} TypeError: unhashable type: 'list'
```

```python
# dict 데이터 조회
print(person2["name"])
print(person2["age"])
# print(person2["city"]) KeyError: 'city'
```

```python
# get 메서드를 활용한 조회
print(person2.get("name"))
print(person2.get("email", "이메일이 존재하지 않습니다."))
```

```python
# 예제
user_data = {"username" : "Doohee", "email" : "s2_pray4u_2s@naver.com", "password" : "1234"}
key = input("확인할 정보를 입력해주세요 (username, email, password): ")
result = user_data.get(key, "존재하지 않는 데이터 입니다.")
print(result)
```

```python
# 데이터 추가 및 수정

# 1) 기본적인 추가, 수정 방법
user_data["phone"] = "010-1234-5678"
user_data["username"] = "WDH"
print(user_data)
```

```python
# 2) update 메서드 활용
user_data.update({"nickname" : "tube"})
print(user_data)

# 키가 문자열인 경우
user_data.update(phone = "010-2345-6789")
print(user_data)
```

```python
# 다른 딕셔너리 추가
extra_data = {"city" : "Seoul"}
user_data.update(extra_data)
print(user_data)
```

```python
# 데이터 삭제
del user_data["city"]
print(user_data)

# 키로 제거
nickname = user_data.pop("nickname")
print("pop >>", user_data, nickname, sep = " /// ")
```

```python
# 가장 마지막 요소 제거
phone = user_data.popitem()
print(user_data, phone, sep = " /// ")
```

```python
# dict 비우기
user_data.clear()
print(user_data)

# dict 삭제하기
del user_data
# print(user_data) >> NameError: name 'user_data' is not defined
```

```python
# 메서드
user_data = {"username" : "Doohee", "email" : "s2_pray4u_2s@naver.com", "password" : "1234" }

# keys: 모든 키를 반환
print("key:", user_data.keys()) # key: dict_keys(['username', 'email', 'password'])
print("key:", list(user_data.keys())) # key: ['username', 'email', 'password']

# values: 모든 값을 반환
print("value:", list(user_data.values())) # value: ['Doohee', 's2_pray4u_2s@naver.com', '1234']

# items: 모든 키, 값 쌍을 반환
print("items:", list(user_data.items())) # items: [('username', 'Doohee'), ('email', 's2_pray4u_2s@naver.com'), ('password', '1234')]
```

```python
# 실습1. 딕셔너리 종합 연습 문제

# 문제1. 딕셔너리 핵심 개념 통합 실습
user = dict()
user.update({"username" : "Skywalker", "email" : "sky@example.com", "level" : 5})
print(user)

email_value = user.get("email")
print(email_value)

user.update(level = 6)
print(user)

key = input("확인할 정보를 입력해주세요.")
result = user.get(key, "미입력")
print(result)

user.update({"nickname" : "sky"})
print(user)

del user["email"]
print(user)

user.setdefault("signup_data", "2025-11-17")
print(user)
```

```python
# 문제2. 학생 점수 관리
students = dict()
students.update({"Alice" : 85, "Bob" : 90, "Charlie" : 95})
print(students)

students["David"] = 80
print(students)

students["Alice"] = 88
print(students)

del students["Bob"]
print(students)

```
