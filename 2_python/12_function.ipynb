```python
'''
함수 (function)
- 특정 작업을 수행하는 코드들의 모음
- 복잡한 코드를 작은 단위로 나눌 수 있게 도와줌
- 특정한 코드들을 재사용 할 수 있게 함.
'''
```

```python
# 사용자 정의 함수 기본 문법
# 함수의 정의: define의 약자로 def 사용

def 함수이름 (매개변수):
    # 실행할 코드
    print(매개변수)
    return "반환값"

# 함수의 실행 (호출 call)
result = 함수이름("인자")
print(result)

# 매개변수 (parameter) : 매개 + 변수
# 매개: 둘 사이를 연결해줌
# 함수가 실행될 때 인자로부터 입력되는 값을 함수의 코드 블록으로 전달하는 역할
```

```python
# 함수의 필요성 예제 

a = 10
b = 20

if a > b:
    print(a - b)
else:
    print(a + b)

c = 30
d = 40

if c > d:
    print(c - d)
else:
    print(c + d)


def my_func(a, b):
    if a > b:
        return a - b
    else:
        return a + b
    
print(my_func(10, 20))
print(my_func(30, 40))
```

```python
# 실습1. 사칙연산 계산기 함수 만들기

# 문제1. 사칙연산 계산기 함수 만들기

a = int(input("숫자를 입력해주세요.:"))
b = int(input("숫자를 입력해주세요.:"))
operator = input("연산자를 입력해주세요. (+, -, *, /):")

def calculate(a, b, operator):
    if operator == "+":
        print(f"{a} + {b} = {a + b}")
    elif operator == "-":
        print(f"{a} - {b} = {a - b}")
    elif operator == "*":
        print(f"{a} * {b} = {a * b}")
    elif operator == "/":
        print(f"{a} / {b} = {float(a / b)}")
    else:
        print("지원하지 않는 연산입니다.")

print(calculate(a, b, operator))
```

```python
# 키워드 인자

#예제1
print("안녕하세요", "반갑습니다.", sep ="-", end = "/")

#예제2
def my_func(a, b, c = None, operator = None):
    if operator == "+":
        return a + b
    else:
        return c
    
print(my_func(10, 20, operator = "+"))


# 기존값 인자
# 단, 기본값 매개변수는 위쪽에 위치해야함.
def greet(name, massage ="안녕하세요~!"):
    print(f"{name}님, {massage}")


# 호출 시 인자 생략 -> 기본값 사용
greet("wdh")
greet("ff", "반갑습니다!")

# 위치 가변 인자
# 여러개의 값을 유돌적으로 받을 수 있음
# 값이 튜플 형대로 받아짐

def add_all(*args):
    return sum(args)

print(add_all(1, 2, 3, 4, 5))

# 키워드 가변 인자
# 여러 키워드 인자를 유동적으로 받을 수 있음
# 딕셔너리 형태로 값이 입력됨

def print_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}:{value}")

print_info(name = "Doohee", age = 20, city = "Seoul")
```

```python
# 실습2. 가변인자 연습하기

# 문제1. 숫자 여러개의 평균 구하기

def average(*args):
    if len(args) == 0:
        return "입력값이 없습니다."
    return sum(args) / len(args)

print(average(1, 2, 3, 4))
```

```python
# 문제2. 가장 긴 문자열 찾기 (max 함수에 대해 찾아보고 풀기)

def my_text(*args):
    if len(args) == 0:
        return "입력값이 없습니다"
    return max(args, key = len)

print(my_text("apple", "banana", "kiwi", "melon"))

def my_text(*args):
    answer = ""
    for s in args:
        if len(s) > len(answer):
            answer = s
    return answer

print(my_text("apple", "banana", "wetermelon", "grape"))
```

```python
# 문제3. 사용자 정보 출력 함수

def info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}:{value}")

info(name = "Doohee", age = 38, email = "s2pray4u2s@naver.com")
```

```python
# 문제4. 할인 계산기

def my_items(**kwargs):
    for key, value in kwargs.items():
        dis = value * 0.9
        print(f"{key}: 할인가 {dis}원, 원가 {value}원")

my_items(fruits = 5000, snack = 2000, drink = 2500)
```

```python
# 전역변수: 함수 밖에 선언된 함수
# 지역변수: 함수 안에 선언된 함수

# 전역변수
x = 200

# 예제
def my_func():
    # 지역변수
    x = 10
    print(x)

my_func()
print("함수 밖:", x)
```

```python
# 예제2
x = 10

def my_func2():
    x = 20
    x += 5
    print("지역변수:", x)

my_func2()

print("전역변수:", x)
```

```python
# 예제3 (global 사용 예제)
x = 10

def my_func3():
    global x # 전역변수 사용 선언
    x += 5
    print("지역변수:", x)

my_func3()

print("전역변수:", x)
```

```python
# 권장되는 패턴
# 부수효과 (side effect)를 발생시키지 않는 함수를 위주로 프로그래밍

x = 10

def my_func4(x):
    x += 5
    return x

x = my_func4(x)

print("전역변수:", x)
```

```python
# 실습3. 전역변수 연습하기

# 문제1. 로그인/로그아웃전역상태관리

current_user = None

def login(name):
    global current_user
    
    if current_user is not None:
        print("이미 로그인 되었습니다.")
    else:
        current_user = name
        print(f"{name}님 로그인 성공")

def  logout():
    global current_user
    
    if current_user is None:
        print("로그인 상태가 아닙니다.")
    else:
        print("로그아웃 하셨습니다.")
        current_user = None


login("Ian")
login("CodingOwl")
logout()
login("CodingOwl")
print(current_user)

```
