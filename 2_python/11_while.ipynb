```python
# 문제2. 업다운 게임

secret_number = 95
count = 0

while True:
    number = int(input("1 ~ 100 사이 숫자를 입력해주세요."))
    count += 1
    if number == secret_number:
         print(f"{count}번 만에 정답을 맞췄습니다.")
         break
    elif number > secret_number:
        print(f"{number}보다는 작아요.")
    else:
        print(f"{number}보다는 커요")
```

```python
# 문제2.
import random

answer = random.randint(1, 100)
num = 0 # 사용자 입력 받을 변수
time = 0 # 실행 횟수를 저장할 수 있는 변수

while num != answer:
    num = int(input("1 ~ 100 사이의 수를 입력해주세요."))
    time += 1

    if num > answer:
        print(f"정답이 {num}보다는 작아요.")
    elif num < answer:
        print(f"정답이 {num}보다는 커요.")

print(f"{time}번 만에 정답을 맞췄습니다.") 
```

```python
# 문제2. 유효한 나이만 평균 내기
times = 0
sum_age = 0

while True:
    age = int(input("나이를 입력해주세요."))
    
    if age >= 120 or age <= 0:
        continue
    
    sum_age += age
    times += 1

    if times == 5:
        break
print(f"총 나이 합계는 {sum_age}, 사람 수는 {times}입니다. 나이의 평균은 {sum_age / times} 입니다.")
```

```python
times = 0
sum_age = 0

while times < 5:
    age = int(input("나이를 입력해주세요."))

    if age > 120 or age <= 0:
        continue

    times += 1
    sum_age += age

print(f"총 나이 합계는 {sum_age}세, 나이 평균은 {sum_age / times}세 입니다.")
```

```python
# 중첩 while 문

# 예제

dan = 2
while dan <= 9:
    num = 1
    print(f"[{dan}단]")

    while num <= 9:
        num += 1
        if num % 2 != 0:
            break
        else:
            print(f"{dan} × {num} = {dan * num}")

    print()
    dan += 1
```

```python
my_id = "codingon"
my_pw = "abc123"

while True:  # 전체 프로그램 반복

    print("=== 로그인 화면 ===")
    print("1. 로그인")
    print("2. 종료")
    choice = input("선택: ")

    # 종료 선택하면 전체 프로그램 종료
    if choice == "2":
        print("프로그램을 종료합니다.")
        break

    # 로그인 선택
    if choice == "1":
        # ID 검사
        while True:
            user_id = input("ID : ")
            user_pw = input("PW : ")

            if user_id == my_id and user_pw == my_pw:
                print("로그인 성공!\n")
                break
            else:
                print("로그인 실패! 다시 시도해주세요.\n")

        # 로그인 성공 후 메뉴 반복
        while True:
            print("===== 메뉴 =====")
            print("1. 정보 보기")
            print("2. 설정")
            print("3. 로그아웃")
            menu_select = input("메뉴 선택: ")

            if menu_select == "1":
                print("회원 정보입니다.\n")

            elif menu_select == "2":
                print("설정 메뉴입니다.\n")

            elif menu_select == "3":
                print("로그아웃합니다.\n")
                break  # 메뉴 while 탈출 → 로그인 화면으로 돌아감

            else:
                print("잘못된 입력입니다.\n")
```

```python
# 실습3. 

# 문제1. 로그인 시스템 구현(1)

# 아이디와 비밀번호를 저장할 변수
user_id = "s2pray4u2s"
user_pw = "wdh1234"

# ID 먼저 검사
while True:
    id_input = input("ID를 입력해주세요.:")

    if id_input != user_id:
        print("ID가 일치하지 않습니다.")
        continue

    # 비밀번호 검사
    while True:
        pw_input = input("비밀번호를 입력해주세요.:")

        if pw_input != user_pw:
            print("비밀번호가 일치하지 않습니다.")
            continue

        print("로그인 성공")
        break
    break
```

```python
# 문제2. 로그인 시스템 구현(2)
# ID, PW 동시에 받는 경우

user_id = "s2pray4u2s"
user_pw = "wdh1234"

while True:  # 전체 프로그램 반복
    print("=== 로그인 화면 ===")
    print("1. 로그인")
    print("2. 종료")
    main_sel = input("선택.:")

    if main_sel == "2":
        print("프로그램을 종료합니다.")
        break
    elif main_sel != "1":
        print("잘못 선택 하셨습니다. \n")
        continue

    # 로그인
    id_input = input("ID:")
    pw_input = input("PW:")

    if id_input != user_id or pw_input != user_pw:
        print("로그인 실패! 다시 시도해주세요. \n")
        continue

    print("로그인 성공! \n")

    # 로그인 후 메뉴 화면
    while True:
        print("=== 메뉴 ===")
        print("1. 정보보기")
        print("2. 설정")
        print("3. 로그아웃")
        print("==============")

        sel = input("메뉴 선택해주세요.")

        if sel == "1":
            print("회원 정보 입니다. \n")
            continue

        elif sel == "2":
            print("설정 메뉴 입니다. \n")
            continue

        elif sel == "3":
            print("로그아웃 입니다. \n")
            break
        
        else:
            print("잘못 입력 하셨습니다. \n")
```
