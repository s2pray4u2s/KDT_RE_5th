```python
# 클래스 기본 문법
# 클래스의 정의
class ClassName:
    # 생성자(constructor): 인스턴스(객체)가 생성될 때 호출
    # 인스턴스 변수를 초기화, 기본 상태 설정
    # 하나의 클래스에서 하나만 정의 가능
    def __init__(self, name):
        # 인스턴스 변수
        # self: 인스턴스 자기 자신을 가리킴
        self.name = name
        self.age = 0

    # 인스턴스 메서드
    def method_name(self):
        print(f"이 인스턴스의 이름은 {self.name} 입니다.")

# 인스턴스 생성
my_instance = ClassName("Doohee")
print(my_instance.name)
my_instance.method_name()

another_instance = ClassName("IU")
another_instance.method_name()
```

```python
# 실습1. class 기본 문법 연습

# 문제1. 책 클래스 만들기
class Book:
    def __init__(self, title, author, total_pages, current_page):
        self.title = title
        self.author = author
        self.total_pages = total_pages
        self.current_page = current_page
    
    def read_page(self, pages):
        
        if self.current_page + pages > self.total_pages:
            self.current_page = self.total_pages
            print("책을 다 읽었습니다.")
        else:
            self.current_page += pages
        
        print(f"책 제목은 {self.title} 입니다.")
        print(f"작가는 {self.author} 입니다.")
        print(f"현재 페이지는 {self.current_page}페이지 입니다.")
    
    def progress(self):
        percent = (self.current_page / self.total_pages) * 100
        print(f"현재 읽은 분량: {percent:.1f}% 입니다")

my_book = Book("Harry Potter", "J. K. Rowling",223 ,20)
my_book.read_page(50)
my_book.progress()
```

```python
class Book:
    def __init__(self, title, author, total_pages, current_page):
        self.title = title
        self.author = author
        self.total_pages = total_pages
        self.current_page = current_page

    def read_page(self, pages):
        self.current_page += pages

        if self.current_page > self.total_pages:
            self.current_page = self.total_pages
            print("책을 끝까지 다 읽었습니다.")

    def progress(self):
        percent = (self.current_page / self.total_pages) * 100
        print(f"현재 읽은 분량: {percent:.1f}% 입니다")

book1 = Book("넥서스", "유발 하하리", 500, 1)
book1.read_page(50)
book1.progress()
```

```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height
    
    def area(self):
        return self.width * self.height
    
w = int(input("가로 길이를 입력해주세요."))
h = int(input("세로 길이를 입력해주세요."))

rect = Rectangle(w, h)

print(f"가로의 길이는 {w}m 입니다.")
print(f"세로의 길이는 {h}m 입니다.")
print(f"사각형의 넓이는 {rect.area()}㎡ 입니다.")
```

```python
# 클래스 변수
# - 클래스가 가지고 있는 변수
# - 모든 인스턴스가 공유할 수 있음

class Dog:
    # 클래스 변수
    kind = "댕댕이"

    def __init__(self, species, name, age):
        self.species = species
        self.name = name
        self.age = age

dog1 = Dog("웰시코기", "리치", 5)
dog2 = Dog("레브라도 리트리버", "호떡", 3)

print("instance1", dog1.kind)
print("instance2", dog2.kind)
print("class", Dog.kind)
```

```python
# 클래스 매서드
# 클래스 자체를 대상으로 동작하는 메서드
# 클래스 데이터를 조작하는데 사용

class Book2:
    book_count = 0

    def __init__(self, title, author):
        Book2.book_count += 1
        self.title = title
        self.author = author
    
    # 클래스 메서드
    @classmethod # 데코레이터
    def get_count(cls):
        print(f"현재 {cls.book_count}권의 책을 소유하고 있습니다.")

book1 = Book2("B1", "author1")
book2 = Book2("B2", "author2")
book3 = Book2("B3", "author3")

Book2.get_count()
        
    
```

```python
# 정적메서드
# - 클래스나 인스턴스의 데이터를 조작하지 않는 클래스 함수
# - 클래스나 인스턴스의 상태에 의존하지 않는 일반 함수
# - 개념적으로는 클래스 와 연관이 있으나, 클래스나 인스턴스의 데이터를 조작하지 않는다.

class OperationTool:

    @staticmethod # 데코레이터
    def add(a, b):
        return a + b

OperationTool.add(10, 20)
```

```python
# 실습2. 클래스 변수, 메서드 연습

# 문제1. User 클래스 구현

class UserClass:
    total_users = 0

    def __init__(self, username):
        UserClass.total_users += 1
        self.username = username
        self.points = 0

    def add_point(self, amount):
        self.points += amount

    def get_level(self):
        
        if self.points >= 500:
            print("당신의 포인트는 Gold 입니다.")
        elif self.points >= 100:
            print("당신의 포인트는 Silver 입니다.")
        else:
            print("당신의 포인트는 Bronze 입니다.")

    @classmethod
    def get_total_users(cls):
        print(f"현재 총 유저 수는 {cls.total_users}명 입니다.")

user1 = UserClass("messi"); user1.add_point(550)
user2 = UserClass("doohee"); user2.add_point(300)

UserClass.get_total_users()
user1.get_level()
user2.get_level()
    
```

```python
# 접근 제어와 정보 은닉
# 데이터 무결성을 보호하기 위함
# 코드 안정성을 향상 시키기 위함

class Person2:
    def __init__(self, name, age):
        # public
        self.name = name
        # private : 언더바 두개 (__)를 변수 앞에 붙여서 정의
        self.__age = age

    # getter
    def get_age(self):
        return self.__age
    
    #setter
    def set_age(self, value):
        if value > 120 or value < 0:
            print("유효하지 않은 나이입니다.")
        else:
            self.__age = value



p1 = Person2("Lee", 15)
print(p1.name)
print(p1.get_age())
p1.set_age(-10)

```

```python
# @property 데코레이터
# - 메서드를 속성처럼 보이게 만들어주는 데코레이터

class Ex:
    def __init__(self):
        self.__value = 0
    
    # getter
    @property
    def value(self):
        return self.__value
    
    # setter
    @value.setter
    def value(self, val):
        if val < 0:
            print("유효하지 않은 값입니다.")
        else:
            self.__value = val

ex1 = Ex()
print(ex1.value)
ex1.value = 100
print(ex1.value)
ex1.value = -100
print(ex1.value)
```

```python
# 실습3. 접근제어 와 정보은닉 연습

# 문제1. UserAccount 클래스: 비밀번호 보호

class UserAccount:

    def __init__(self, username, password):
        self.username = username
        self.__password = password

    def change_password(self, old_pw, new_pw):
        if old_pw == self.__password:
            self.__password = new_pw
            print("비밀번호 변경 하였습니다.")
        else:
            print("비밀번호 불일치!")

    def get_password(self):
        return self.__password
    
    def check_password(self, password):
        return password == self.__password
    
user1 = UserAccount("doohee", "wdh1234")
print(user1.username)
print(user1.get_password())

print(user1.check_password("wdh1234"))
print(user1.check_password("wrong")) 

user1.change_password("wdh1234", "new1234")

print(user1.check_password("wdh1234"))
print(user1.check_password("new1234"))
```

```python
# 문제2. Student 클래스: 성적검증(@property 사용)

class Student:

    def __init__(self, score):
        self.score = score
        
    @property
    def score(self):
        return self.__score
    
    @score.setter
    def score(self, my_score):
        if my_score < 0 or my_score > 100:
            raise ValueError("점수는 0~100 사이만 가능합니다.")
        else:
            self.__score = my_score

Student1 = Student(90)
print(f"해당 학생은 {Student1.score}점 입니다.")

Student2 = Student(150)
print(f"해당 학생은 {Student2.score}점 입니다.")
```

```python
class Student:

    def __init__(self, name, score):
        self.__score = score
        self.name = name

    @property
    def score(self):
        return self.__score
    
    @score.setter
    def score(self, my_score):
        if 0 <= my_score <= 100:
            self.__score = my_score  
        else:
            raise ValueError("점수는 0~100 사이만 가능합니다.")

s1 = Student("Alice", 90)
print(f"해당 학생이름은 {s1.name} 입니다.")
print(f"해당 학생은 {s1.score}점 입니다.")

s1.score = 150
print(s1.score)
```

```python
# 상속
# 부모 클래스의 속성과 메서드를 물려받아 새로운 자녀 클래스를 만드는 것
# 상속 기본 문법
# 부모 클래스
class Animal:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def bark(self):
        print("동물이 울음소리를 냅니다.")

#자녀 클래스
class Dog(Animal):
    def __init__(self, name, age, species):
        # super() 는 부모를 가리킴
        super().__init__(name, age)
        self.species = species

    # 오버라이딩
    def bark(self):
        print("왈왈")

dog = Dog("호떡", 5, "레브라도 리트리버")
dog.bark()
print(dog.name)
print(dog.age)
print(dog.species)
```

```python
# 실습4. Shape 클래스 오버라이딩

class Shape:
    def __init__(self, sides, base):
        self.sides = sides
        self.base = base
    def printInfo(self):
        print(f"변의 개수: {self.sides}")
        print(f"밑변의 길이: {self.base}")
    def area(self):
        print("넓이 계산이 정의되지 않습니다.")

class Rectangle(Shape):
    def __init__(self, sides, base, height):
        super().__init__(sides, base)
        self.height = height
    def area(self):
        print(f"{self.sides}각형의 넓이는 {self.base * self.height}㎡ 입니다")

class Triangle(Shape):
    def __init__(self, sides, base, height):
        super().__init__(sides, base)
        self.height = height
    def area(self):
        print(f"{self.sides}각형의 넓이는 {(self.base * self.height) / 2}㎡ 입니다")


shape1 = Shape(4, 10)
shape1.printInfo()
shape1.area()
print("=================================")
rec = Rectangle(4, 10, 5)
rec.printInfo()
rec.area()
print("=================================")
tri = Triangle(3, 10, 5)
tri.printInfo()
tri.area()

        
```

```python
# 추상 클래스 (Abstract Class)
# 클래스의 구조를 정의하는 클래스

from abc import ABC, abstractmethod

class Animal(ABC):
    # 추상메서드
    @abstractmethod
    def bark(self):
        pass

class Dog(Animal):
    def bark(self):
        print("왈왈")

# a = Animal()
a = Dog()
a.bark()
```

```python
# 실습5. 추상클래스 연습문제

from abc import ABC, abstractmethod

class Payment(ABC):
    
    @abstractmethod
    def pay(self, amount):
        pass

class CardPayment(Payment):
    
    def pay(self, amount):
        print(f"CardPayment: 카드로 {amount}원을 결제합니다.")

class CashPayment(Payment):
        
    def pay(self, amount):
        print(f"CashPayment: 현금으로 {amount}원을 결제합니다.")


card = CardPayment()
card.pay(5000)

cash = CashPayment()
cash.pay(12000)


```
