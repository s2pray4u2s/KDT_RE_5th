```python
# 문제1. 슬라이싱으로 특정 구간 뒤집기
data = ["A", "B", "C", "D", "E", "F", "G"]
print(data[2:5][::-1])
```

```python
# 문제2. 여러 변수 입력 심화
location = ["Seoul", "Busan", "Daejeon", "Daegu"]
print(location[0:2])
print(location[2:])
```

```python
# 문제3. 문자열에서 숫자만 뽑기
text = "a1b2c3d4"
print(text[1::2])
```

```python
# 문제4. 삭제 + 이어붙이기
nums = [10, 20, 30, 40, 50, 60]
nums1 = nums[::2]
nums2 = nums1 * 2
print(nums2)
```

```python
# 문제5. 리스트 교체 심화
colors = ["Red", "Green", "Blue", "Yellow", "Black", "Pink"]
colors[2:5] = ["White"]
print(colors)
```

```python
# 문제6. 문자열로 리스트 만들기
data = "10,20,30,40,50"
list = data.split(",")
print(list[:3])
```

```python
# 문제7. 학생 점수 처리
score = "85 92 77 64 100 88"
score1 = score.split(" ")
score2 = [int(score1[0]), int(score1[1]), int(score1[2]), int(score1[3]), int(score1[4]), int(score1[5])]
print(score2)
print(f"최대값: {max(score2)} 점, 최소값: {min(score2)} 점")
```

```python
# 문제8. 영화 데이터 구성
name = ["Interstellar", "Inception", "Dunkirk"]
year = [2014, 2010, 2017]
print(f"영화 {name[0]}는 {year[0]}년에 개봉했다.")
print(f"영화 {name[1]}은 {year[1]}년에 개봉했다.")
print(f"영화 {name[2]}는 {year[2]}년에 개봉했다.")
```

```python
# 문제9. 입력 데이터 구조화
icebreaking = "홍길동 28 INFJ 개발자"
icebreaking1 = icebreaking.split(" ")
print(f"이름: {icebreaking1[0]} / 나이: {icebreaking1[1]} / MBTI: {icebreaking1[2]} / 직업: {icebreaking1[3]}")
```

```python
# 문제10. 중간값만 뽑기
nums = [100, 200, 300, 400, 500]
print(nums[2])
```

```python
# 문제11. 두 리스트 섞어서 출력하기
a = ["A", "B", "C"]
b = [1, 2, 3]
print(f"{a[0]}-{b[0]}, {a[1]}-{b[1]}, {a[2]}-{b[2]}")
```

```python
# 문제12. 특정 규칙 조합 만들기
items = ["x", "y", "z", "k", "m", "n"]
item = items[0:2] + items[4:]
print(item)
```
