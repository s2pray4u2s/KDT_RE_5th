```python
# 문제1. 두 수를 입력받아 더한 값을 출력하라.
A = int(input("숫자1을 입력해주세요."))
B = int(input("숫자2을 입력해주세요."))
print(f"두 숫자의 합은 {A + B} 입니다.")
```

```python
# 문제2. 숫자 리스트가 있을 떄, 각 요소를 3배 한 새 리스트를 만들어라.
nums = [1, 2, 3, 4]
my_nums = [nums[0]*3, nums[1]*3, nums[2]*3, nums[3]*3]
print(my_nums)
```

```python
# 문제13. 세 정수를 입력 받아 평균을 구하라.
num1 = int(input("숫자1을 입력해주세요."))
num2 = int(input("숫자2을 입력해주세요."))
num3 = int(input("숫자3을 입력해주세요."))
numbers = [num1, num2, num3]
print(f"세 숫자의 평균은 {sum(numbers) / len(numbers)}입니다.")
```

```python
# 문제14. 리스트의 모든 요소를 문자열로 변환한 새 리스트 만들기
nums = [10, 20, 30]
nums1 = list(map(str, nums))
print(nums1)
```

```python
# 문제15. 리스트의 모든 요소를 하나의 문자열 "10-20-30" 으로 합쳐라.
nums = [10, 20, 30]
nums1 = list(map(str, nums))
print(f"{nums1[0]}-{nums1[1]}-{nums[2]}")
```

```python
# 문제17. 공백으로 구분된 숫자 "1 2 3 4"를 리스트 [1, 2, 3, 4]로 만들기
nums = "1 2 3 4".split(" ")
print(list(map(int, nums)))
```

```python
# 문제18. 리스트 요소의 합과 개수를 이용하여 평균 구하기
nums = [5, 10, 15, 20]
print(f"숫자의 평균은 {sum(nums) / len(nums)} 입니다")
```

```python
# 문제20. 문자열 리스트에서 길이가 3 이상인 단어만 모아 새 리스트 만들기.
words = ["hi", "python", "go", "sun"]
result = []
if len(words[0]) >= 3:
    result.append(words[0])
if len(words[1]) >= 3:
    result.append(words[1])
if len(words[2]) >= 3:
    result.append(words[2])
if len(words[3]) >= 3:
    result.append(words[3])
print(result)
    
```
